```python
import pandas as pd
df = pd.read_csv("new_data.csv")
df = df.drop("Unnamed: 0", axis=1)
df.head()
```

```python
tradelane_groups = df.groupby('Shipments: Attributes Tradelane')
```

```python
df['Shipments: Attributes Tradelane'].value_counts()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, r2_score

results = {}

for tradelane, group in tradelane_groups:
    print(f"\nTraining model for tradelane: {tradelane}")
    
    if group.shape[0] < 50:
        print(f"Skipping {tradelane} due to insufficient data ({group.shape[0]} rows)")
        continue
    
    # Drop shipment ID
    group = group.drop("Shipments: Attributes Shipment ID", axis=1, errors='ignore')
    
    # Prepare X and y
    X = group.drop(["Total Actual Days", "Shipments: Attributes Tradelane"], axis=1)
    X = pd.get_dummies(X)  # one-hot encoding
    y = group["Total Actual Days"]
    
    # Train/test split
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
    
    # Train model
    model = RandomForestRegressor(
        n_estimators=500, max_features='log2',
        min_samples_split=10, random_state=0)
    model.fit(X_train, y_train)
    
    # Predictions
    preds = model.predict(X_test)
    
    # Evaluation
    mae = mean_absolute_error(y_test, preds)
    r2 = r2_score(y_test, preds)
    
    results[tradelane] = {"MAE": mae, "R2": r2, "Samples": len(group)}
    print(f"Tradelane: {tradelane} | MAE: {mae:.2f} | R²: {r2:.2f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

feature_importances = pd.Series(model.feature_importances_, index=X_train.columns)
top_features = feature_importances.sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_features.values, y=top_features.index)
plt.title("Top 10 Important Features for TPEB Tradelane")
plt.xlabel("Feature Importance")
plt.ylabel("Features")
plt.tight_layout()
plt.show()
```
