```python
# dummy_df = pd.get_dummies(lcl_df, dtype=int)
# dummy_df.head()
numerical_coln = ['Shipments: Attributes Gross Weight (kg)',
                  'Shipments: Attributes Volume (cbm)',
                  'Latitude',
                  'Longitude',
                  'Shipment Containers: Distance Final Port to Destination Distance in mile',
                  'Shipment Containers: Distance Origin to First Port Distance in km']
categorical_coln = ['Shipments: Attributes Ocean type category',
                   'Shipments: Attributes Tradelane',
                   'Port pair']
lcl_df = lcl_df.drop("Shipments: Attributes Shipment ID", axis=1)
```

```python
# from sklearn.preprocessing import StandardScaler


# scaler = StandardScaler()
# scaled_df = pd.DataFrame(scaler.fit_transform(lcl_df[numerical_coln]), columns=numerical_coln)

# dummies_df = pd.get_dummies(lcl_df[categorical_coln], dtype=int)

# target_df = lcl_df["Total Actual Days"]
# # Concatenate the scaled numerical data and dummy variables
# final_df = pd.concat([scaled_df, dummies_df, target_df], axis=1)

final_df = lcl_df.dropna()
final_df = pd.get_dummies(final_df)
```

```python
X = final_df.drop("Total Actual Days",axis=1)
y = final_df["Total Actual Days"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf_model = RandomForestRegressor(n_estimators=500,
                                max_depth=None,
                                max_features='log2',
                                min_samples_leaf=1,
                                min_samples_split=10,
                                random_state=0)
rf_model.fit(X_train, y_train)
```

```python
predictions = rf_model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error, r2_score

mae = mean_absolute_error(y_test, predictions)
mse = mean_squared_error(y_test, predictions)
r2 = r2_score(y_test, predictions)

print(f"Mean absolute Error: {mae}")
print(f"Mean Squared Error: {mse}")
print(f"R^2 Score: {r2}")
```
